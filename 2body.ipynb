```python
def cm_u(r):
    return r*(1e-12)/(5.11)

def CM_12(m1,m2,r):
    return m2/(m1+m2)*r, -m1/(m1+m2)*r
```

```python
a_cm=1.5e13
e=0.0167
theta0=0
m1=1e-6
m2=1
```

```python
a=cm_u(a_cm)
print(a)
T=2*np.pi*np.sqrt(a**3/m2)
print(T)
```

```python
M=m1+m2
mu=m1*m2/M
print(M, mu)
```

```python
l=np.sqrt(M*a*(1-e**2))
r_p=a*(1-e)
r_a=a*(1+e)
theta_p=theta0
print(r_p,theta_p)
#r_a=a/(1-e**2)
print(r_a)
print(l)
```

```python
r_p=l**2/((e*np.cos(theta_p)+1)*M)
print(r_p)
```

```python
E=-2*M/(2*a*mu)
print(E)
```

```python
v_theta_p=l/r_p


#v_r_p=np.sqrt(-v_theta_p**2+2*M/r_p-M/(a))
#print(-(l/r_p)**2+2*M/r_p-M/a/mu)
#print(v_r_p,v_theta_p)

v_p=np.sqrt(M*(2/r_p-1/a))

v_r_p=np.sqrt(v_p**2-v_theta_p**2)
v_r_p=0
#print(v_p-v_theta_p)
print(v_r_p, v_theta_p, v_p)
```
